### Task 1: 
Event detection: Identify the most prominent event. There is a very significant drop (>3-sigma) in all time series. Identify it, figure out the date (you know when the data starts and what the cadence is) and figure out what it is due to.

```python
# summing all rides for all stations and card types
total_ridership = np.nansum(data, axis=(0,1))
print (total_ridership.shape)
```

```python
# defining 3 sigma threshold range
mean_tr = np.nanmean(total_ridership)
std_tr = np.nanstd(total_ridership)
threshold = [mean_tr + 3 * std_tr, mean_tr - 3 * std_tr]
threshold
```

```python
# defining a dataframe
df_tr = pd.DataFrame(data=total_ridership, index=date, columns=['rides']).reset_index()
df_tr.columns = [['date', 'rides']]
df_tr.head()
```

```python
#Visualising the data
ax = pl.figure(figsize=(10,6)).add_subplot(111)
ax.plot(df_tr.date, (df_tr.rides), color = 'black', alpha = .95, label = 'Rides')
ax.plot([df_tr.date[0], df_tr.date[193]], [threshold[1], threshold[1]], color = 'orange', label = 'Lower 3 Sigma Threshold')
ax.plot([df_tr.date[0], df_tr.date[193]], [threshold[0], threshold[0]], color = 'red', label = 'Higher 3 Sigma Threshold')
ax.plot([df_tr.date[0], df_tr.date[193]], [mean_tr, mean_tr], '--', color = 'green', label = 'Mean')
ax.set_title('Weekly MTA rides', fontsize=14)
ax.set_ylabel('Number of rides in 10 millions', fontsize=12)
ax.set_xlabel("Date", fontsize=12)
ax.legend(loc=4)
labels = ax.get_xticklabels()
pl.setp(labels, rotation=45, fontsize=10)
pl.show()
```

#### Figure 1: The graph shows total number of rides in MTA per week for period starting 5/21/2010. Graph additionally provides the inforamtion regarding the mean and 3-sigma threshold for the period. It is noted that most of the weeks number of rides are within the 3-sigma threshold, except for a event which occures approximately in October-November of 2012.

### Task 2: 
#### Some of the time series are stationary, some have a complex structure, some show a downward trend: Identify the **ridership types**  that have steadily increased in popularity,  and that have steadily decreased. by how much? (e.g what is the ratio of usage in the first 10 and last 10 weeks)

```python
# defining card types
ridetype = np.array(['10t', '14d', '1d', '2t', '30d', '7d', 'ada', 'afas', 'exp','ez',
            'ff','mc','mr','month','rfm','rr','sen','spec','tcmc','t',
            'umlim','xbus','stud'])
```

```python
#summing all card types by week
card_sum = np.nansum(data, axis=0)
card_sum.shape
```

```python
# saving the information into a pandas dataframe
df_card = pd.DataFrame(data=card_sum.T, index=date, columns=ridetype)
df_card.head()
```

```python
# calculating rolling means
hg = pd.rolling_mean(df_card, 10)
hg.head(11)
```

```python
# runnning for-loop to determine trend 
y = [] # to collect shortlisted details of shortlisted card types
y1 = [] # to collect shortlisted details of all card types
time = np.arange(194) # defining time range
for column in hg:
    d = pd.DataFrame(hg[column]).reset_index().reset_index() # defining a dataframe with data for just one card-type 
    d.columns = [['a', 'b', 'c']] # renmaning columns for ease of operations
    x = (d['c'][193]) / (d['c'][9]) # calculating ratio of first and last 10 weeks
    res = smf.ols(formula = 'c ~ a', data = d).fit()  #running linear regression
    y1.append({'card':column, 'slope':res.params[1], 'r2':res.rsquared_adj, 'change ratio':x, 
                  'change rate': ((d['c'][193] - d['c'][9]) / 194)}) # recording the results
    if (res.rsquared_adj > .75) & (x > 1.05 or x < .95): # weeding out outliers and no trend
        y.append({'card':column, 'slope':res.params[1], 'r2':res.rsquared_adj, 'change ratio':x, 
                  'change rate': ((d['c'][193] - d['c'][9]) / 194)}) # recording shortlised
        # ploting the shortlised card types which show significant trend
        ax = pl.figure(figsize=(10,5)).add_subplot(111)
        ax.plot(d['c'], color='black', label = 'rides')
        ax.plot( res.predict(), color='red', label = 'line fit')
        ax.set_title('Card type = %s, R2_adj = %s, Change Ratio = %s'%(column, res.rsquared_adj, x),
                     fontsize = 16)
        ax.set_ylabel("Number of rides", fontsize=14)
        ax.set_xlabel("Weeks", fontsize=14)
        ax.legend(loc=4)

df_trend =   pd.DataFrame(y1).sort_values(by='change ratio')      
df_results = pd.DataFrame(y).sort_values(by='change ratio')
print ('The table of ridership trend for each card type sorted by the ratio of change in ridership between first and last 10 weeks')
print (df_trend)
print ()
print ('The table of significant trend filtered by adjested r-squared to weed out structual change induced abnormalities')
(df_results)
```

#### Figure 2: Card type Month has most negative trend of all cards with change ratio of 0.58 and adjusted R-square of 0.87, while card type SEN has the most positive trend with change ratio of 2.90 and adjusted R-square of 0.96. Card types rr, 7d and 14d also have significantly positive trend. There are card types other than month type which exhibit fall in ridership but these were deemed not tread but impact of a structural change as their adjusted R-square values were very low.

### Task 3:
#### Several stations show a prominent annual periodicity. Identify the 4 **stations** (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). (Can you figure out what the periodic peak in rides is due to?)

```python
#summing all stations by week
station_sum = np.nansum(data, axis=1)
station_sum.shape
```

```python
# saving the information into a pandas dataframe
df_station = pd.DataFrame(data=station_sum.T, index=date)
df_station.head()
```

```python
# performing fourier transformation on all stations and 
# recording INDEX 4 of fourier tranformation as the power value closest to an annual period is at INDEX 4

fourier = []
for column in df_station:
    ft = np.abs(np.fft.rfft(df_station[column]))
    fourier.append({'station_id': column, 'yr_periodicity_score': ft[4]})
    
df_ft = pd.DataFrame(fourier).sort_values(by='yr_periodicity_score', ascending=False)
(df_ft.head(6))
```

#### Stations 0, 194, 150 and 327 are the top four stations with high annual periodicity. 

```python
# Visualizing the results 
for column in [0,194,150,327]:
    ft = np.abs(np.fft.rfft(df_station[column]))
    N = len(df_station)
    # ploting rides
    fx = pl.figure(figsize=(15,5)).add_subplot(111)
    fx.plot(df_station[column], color='black', label='rides')
    fx.set_title('MTA ridership pattern for station id %s'%(column), fontsize = 20)
    fx.set_ylabel("Number of rides", fontsize=20)
    fx.set_xlabel("Weeks", fontsize=20)
    fx.legend(loc=4)
    # ploting power spectrum from furnier transformation
    ax=pl.figure(figsize=(15,5)).add_subplot(111)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (ft[1:]), 'bo', ms=10)
    ax.plot(np.fft.rfftfreq(N, 1.0)[4], (ft[4]), 'ro', ms=20, label='52nd week')
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (ft[1:]), color ='black', label = 'power')
    ax.get_xticks(minor=True)
    ax.set_xticklabels([ "%s"%(1/f) for f in ax.get_xticks()], fontsize=20)
    ax.legend(loc=1)
    pl.xlabel("period (week)", fontsize=20)
    pl.ylabel("Power", fontsize=20)
    pl.title("MTA ridership power spectrum for station id %s"%(column), fontsize=20)
```

#### Figure3: Stations 0, 194, 150 and 327 are the top four stations with high annual periodicity. Though it is noted that the annual periodicity can be easily (visually) noted in the ridership plot of the stations 194, 150 and 327 without refering the furnier transformation but same doesn't apply to station 0 even though it has the highest power value.

### For station 0: It is uncertain as to what is the cause of the periodicity.
### For station 194: The periodicity seems to be seasonal, more rides in summer and lesser rides in winter. Unable to spot an annual event.
### For station 150: The periodicity seems to be seasonal, more rides in summer and lesser rides in winter. But there seems to be some annual event aroud June-July.
### For station 327: The periodicity seems to be linked to some particular event that takes place in September.

## EXTRA CREDIT: Cluster. 
### Cluster the time series, i.e. identify groups of time series that look similar among each other. You can use KMeans for example to identify common trends, or PCA (Principle Component Analysis). 

```python
# using Federica's code from instructions
tots = data.transpose(2,0,1).reshape(data.shape[2], data.shape[1]*data.shape[0]).T
tots = tots[tots.std(1)>0]
km = KMeans(n_clusters=8)
vals = ((tots.T - tots.mean(1))/tots.std(1)).T
km.fit(vals)
cluster_label = km.predict(vals)
```

```python
# running PCA analysis
n = vals.shape[1]
pca = PCA(n)
Xproj = pca.fit_transform(vals)
eigenvalues = pca.explained_variance_
print (pca.explained_variance_ratio_)
print ('The first and second eigenvectors could explain ', ((pca.explained_variance_ratio_[0] + pca.explained_variance_ratio_[1])), '% of total the variance.')
```

```python
# visualising the clusters using first and second eigenvectors
pl.figure(2, figsize=(10, 10))
pl.clf()
pl.scatter((Xproj[:, 0]), (Xproj[:, 1]), c=cluster_label, cmap=plt.cm.Paired)
pl.title("PCA clustering of the time series", fontsize=14)
pl.xlabel('First Eigenvalue', fontsize=12)
pl.ylabel('Second Eigenvalue', fontsize=12)
pl.show()
```

### Figure 4: The first and second eigenvectors of time series are clustered into 8 clustered using K-Means and PCA.

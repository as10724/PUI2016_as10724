## In log space four of the infrastructure variables have pvalues that indicate they have impact on number of complaints. These are: households with computer, households with broadband, households without internet and households with no computer. Again the result is counter intutive, therefore I believe it would be helpful to run a multi-variant regression for these variables.

```python
#Running multivariant regression
regress15b = smf.ols(formula = 'np.log10(no_of_complaints) ~  np.log10(with_computer) + np.log10(with_broadband) + np.log10(without_internet) + np.log10(no_computer)', data = final3).fit()
regress15b.summary()
```

```python
regressf = smf.ols(formula = 'np.log10(no_of_complaints) ~  np.log10(Total_Population) + np.log10(employed_16) + np.log10(without_internet) -1', data = final3).fit()
regressf.summary()
```

## With R-squared value of the model equal to 0.998, it is concluded the model explains most of the data.

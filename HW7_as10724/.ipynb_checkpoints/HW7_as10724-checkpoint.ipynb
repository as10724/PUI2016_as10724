# HW 7 : Ploting a Graph

```python
#improting required modules
from __future__ import print_function, division
import pandas as pd
import pylab as pl
import os
import json
import seaborn as sns
import statsmodels.formula.api as smf
import scipy.stats
%pylab inline
```

```python
# Importing hourly Air pollution and weather data of Delhi (India) for month of November, 2015. 
# Data is sourced from the website Central Pollution Control Board of India (http://www.cpcb.gov.in/CAAQM/frmUserCompReportCriteria.aspx)
# Data was collected, cleaned and written on to a csv file, which is stored in HW7_as10724 folder and is being recalled here.

data1 = pd.read_csv('PollutionData_Delhi_Nov_2015.csv')
data1.head()
```

```python
# converting time columns to datetime
```

```python
data1['dt'] = pd.to_datetime(data1['Unnamed: 0'] + " " + data1['Unnamed: 1'])
data1['dt'].head()
```

```python
data1.head()
```

```python
#saving the required information into a new dataframe

data = data1[[u'pm25', u'rh', u'at', u'ws', 'dt']]
data.columns
```

```python
#Renaming columns

data.columns = [[u'pm25', u'rh', u'temp', u'ws', 'dt']]
data.columns
```

```python
data.head()
```

```python
type(data.dt.values[0])
```

```python
# Chnaging th color scheme to color blind friendly

current_palette = sns.color_palette('colorblind')
sns.palplot(current_palette)
```

```python
# Carrying out linear regression analysis to establish the impact of wind speed, humidity and temperature on the concentration
# of PM2.5 (particulate matter - type of air pollutent).

Regressd = smf.ols(formula = 'pm25 ~ rh + temp + ws', data = data).fit()

Regressd.summary()
```

 ## Ploting 

```python
#Graph 1
ax = pl.figure(figsize=(10,10)).add_subplot(111)
data.plot(kind='scatter', x= 'ws', y='rh', ax=ax,
                fontsize=22, colormap='RdBu_r', alpha = .75, 
                marker='o',  figsize=(16, 14),
                s=(data['pm25'] * 2), 
                c=data['temp'])
ax.set_title('Delhi Air Pollution (Color maps:Temperature in C; Size maps: Hourly PM2.5 concentration)', fontsize=22)
ax.set_ylabel("Relative humidity (%)", fontsize=20)
ax.set_xlabel("Wind speed (m/s)", fontsize=20)
```

## Figure 1: Impact of weather elements on the Winter air pollution levels in Delhi. The graph shows that the concentration of PM2.5 (bubble size) increases when wind speed is low, relative humidity is high and temperature is low. This is in accordance with the results of multivarient linear regression analysis carried out in this study.

```python
#Graph 2
ax = pl.figure(figsize=(10,10)).add_subplot(111)
data.plot(kind='scatter', x= 'temp', y='pm25', ax=ax,
                fontsize=22, colormap='gist_heat_r', alpha = .95, 
                marker='o',  figsize=(16, 14),
                s=(data['ws'] * 500), 
                c=data['rh'])
ax.set_title('Delhi Air Pollution (Color maps:Relative Humidity; Size maps: wind speed)', fontsize=20)
ax.set_ylabel("Hourly PM2.5 concentration (microgram/ m3)", fontsize=18)
ax.set_xlabel("Temperature in Celsius", fontsize=18)
```

## Figure 2: Impact of weather elements on the Winter air pollution levels in Delhi. The graph shows that the concentration of PM2.5 (y-axis) increases when wind speed(bubble size) is low, relative humidity (color) is high and temperature is low (a-axis). This is in accordance with the results of multivarient linear regression analysis carried out in this study.

```python
ax = pl.figure(figsize=(10,10)).add_subplot(111)
data.plot(kind='scatter', x= 'temp', y='pm25', ax=ax,
                fontsize=22, colormap='gist_heat_r', alpha = .95, 
                marker='o',  figsize=(16, 14),
                s=(data['rh'] * 10), 
                c=data['ws'])
ax.set_title('Delhi Air Pollution (Color maps:wind speed; Size maps: Relative Humidity)', fontsize=20)
ax.set_ylabel("Hourly PM2.5 concentration (microgram/ m3)", fontsize=18)
ax.set_xlabel("Temperature (Celsius)", fontsize=18)
```

## Figure 3: Impact of weather elements on the Winter air pollution levels in Delhi. The graph shows that the concentration of PM2.5 (y-axis) increases when wind speed(color) is low, relative humidity (bubble size) is high and temperature is low (x-axis). This is in accordance with the results of multivarient linear regression analysis carried out in this study.

```python
#Graph 4
data.plot(x='dt', y='pm25',
                fontsize=16, colormap='RdBu', alpha = .95, 
                figsize=(16, 14),
                legend=True)
pl.title('Delhi Air Pollution (hourly concentration of PM2.5)', fontsize=20)
pl.ylabel("Hourly PM2.5 concentration (microgram/ m3)", fontsize=18)
pl.xlabel("Date Time", fontsize=18)
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pylab as pl
from geopandas import GeoDataFrame
from fiona.crs import from_epsg
import shapely as shp
import seaborn 
import os
import geopandas as gpd

print (os.getenv("PUIDATA"))
```

1.. Find the census trck shape files online (American Fact Finder has them), http://www1.nyc.gov/ has them. The folder you should download is names nycb2010_16d. 
Load it as a geopandas file

```python
#importing and unzipping census track file to PUIdata
os.system('curl -o ' + os.getenv('PUIDATA') + '/nycb2010_16d.zip' +
          ' http://www1.nyc.gov/assets/planning/download/zip/data-maps/open-data/nycb2010_16d.zip')
os.system("unzip " + os.getenv("PUIDATA") + "/nycb2010_16d.zip -d " + os.getenv("PUIDATA"))
```

```python
#writing the data onto a geopandas dataframe
shape_path = os.getenv('PUIDATA') + '/nycb2010_16d/nycb2010.shp'
ct_shape = GeoDataFrame.from_file(shape_path)
ct_shape.head()
```

```python
#converting the coordinates to lat/lon via epsg=4326
ct_shape = ct_shape.to_crs(epsg=4326) 
```

2.. Extract only the Brooklyn shape polygons and convert the coordinates to lat/lon 

```python
#extracting brooklyn
brooklyn_shape = ct_shape[ct_shape['BoroName'] == 'Brooklyn']
brooklyn_shape.head()
```

```python
#visiualising brooklyn
f,ax = pl.subplots(figsize=(10,8))
brooklyn_shape.plot(alpha=0.5, linewidth=0.5, ax=ax)
ax.set_title("Census Tract map of Brooklyn", fontsize=20)
ax.set_ylabel('Latitude', fontsize=18)
ax.set_xlabel("Longitude", fontsize=18)
```

3.. Find the coordinates of CUSP: 1 MetroTech Brooklyn. Create a Point shapely geometry with the coordinates of CUSP


```python
cusp_latlon = (-73.9885917, 40.6932169) #CUSP: 1 MetroTech Brooklyn coordinates via google
CUSP = pd.DataFrame(['CUSP'], columns=['Location']) #creating a dummy dataframe
cusp_geo = gpd.GeoSeries(shp.geometry.Point(cusp_latlon)) #defining shapely geometry
cusp_location = GeoDataFrame(CUSP, geometry=cusp_geo) #saving cusp location to a geopandas dataframe
cusp_location
```

4.. Loop over the Brookly polygons (nearly 10000) and assess find in which geometry the polygon is contained

```python
#using geo merger to locate CUSP in Brooklyn census track instead of running a loop 
cusp_location.crs = brooklyn_shape.crs #matching crs of the two geodataframes
bklyn_cusp = gpd.sjoin(brooklyn_shape, cusp_location, how="inner", op='contains') #geo merger
bklyn_cusp.head()
```

5.. Plot one every 10 polygons of brooklyn (using the geopandas datafame method plot)
and overplot the locati`on of cusp as a point (using pylab.scatter)
Overplot also the census track that contains cusp in red (using the geopandas datafame method plot)

```python
#visualising the results
f,ax = pl.subplots(figsize=(10,8))
brooklyn_shape.iloc[np.arange(0, len(brooklyn_shape), 10),:].plot(alpha=0.5, linewidth=0.5, ax=ax) #ploting one every 10 polygons
pl.scatter(cusp_latlon[0], cusp_latlon[1], s=300, c='orange', alpha=.5)
bklyn_cusp.plot(ax=ax, color='red')
ax.set_title("CUSP location on Census Tract map of Brooklyn", fontsize=20)
ax.set_ylabel('Latitude', fontsize=18)
ax.set_xlabel("Longitude", fontsize=18)
```
